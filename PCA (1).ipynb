```python
import numpy as np

# Objective function
def f(x):
    return x**2 - 4*x + 4  # Minimum at x = 2

# Parameters
GRID_SIZE = (10, 10)
MAX_ITER = 100
SEARCH_SPACE = (-10, 10)
LEARNING_RATE = 0.5  # how strongly a cell moves toward best neighbor

# Initialize population (random values between -10 and 10)
grid = np.random.uniform(SEARCH_SPACE[0], SEARCH_SPACE[1], GRID_SIZE)

# Evaluate fitness
def evaluate_fitness(grid):
    return f(grid)

# Get 3x3 neighborhood with wrap-around boundaries
def get_neighbors(grid, i, j):
    rows, cols = grid.shape
    neighbors = []
    for di in [-1, 0, 1]:
        for dj in [-1, 0, 1]:
            ni = (i + di) % rows
            nj = (j + dj) % cols
            neighbors.append(grid[ni, nj])
    return np.array(neighbors)

# Main loop
for iteration in range(MAX_ITER):
    fitness_grid = evaluate_fitness(grid)
    new_grid = np.copy(grid)

    for i in range(GRID_SIZE[0]):
        for j in range(GRID_SIZE[1]):
            neighbors = get_neighbors(grid, i, j)
            neighbor_fitness = f(neighbors)

            # Identify best neighbor (lowest fitness)
            best_neighbor = neighbors[np.argmin(neighbor_fitness)]

            # Update rule: move partially toward best neighbor
            new_grid[i, j] = grid[i, j] + LEARNING_RATE * (best_neighbor - grid[i, j])

    grid = new_grid

    # Track best cell
    best_idx = np.unravel_index(np.argmin(f(grid)), grid.shape)
    best_x = grid[best_idx]
    best_fitness = f(best_x)

    if iteration % 10 == 0 or iteration == MAX_ITER - 1:
        print(f"Iteration {iteration+1}/{MAX_ITER} -> Best x: {best_x:.6f}, f(x): {best_fitness:.6f}")

# Final output
print("\nBest solution found:")
print(f"x = {best_x:.6f}")
print(f"f(x) = {best_fitness:.6f}")
```
